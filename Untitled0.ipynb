```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. Chargement du fichier que tu as uploadÃ©
df = pd.read_csv('social_media_addiction 2.csv')

# 2. On dÃ©finit nos variables pour le modÃ¨le
# X = Age et Temps d'utilisation (min) | y = Indice de santÃ© mentale
X = df[['Age', 'Daily_Usage_Time_min']]
y = df['Mental_Health_Index']

# 3. CrÃ©ation et entraÃ®nement de l'IA
model = LinearRegression()
model.fit(X, y)

# 4. ðŸ”® LA PRÃ‰DICTION
# Test : Un Ã©tudiant de 20 ans qui passe 3h (180 min) sur les rÃ©seaux
age_test = 20
temps_min = 180
prediction = model.predict([[age_test, temps_min]])

print(f"--- ðŸŽ¯ RÃ‰SULTAT DU MODÃˆLE ---")
print(f"Pour un profil de {age_test} ans avec {temps_min} min d'Ã©cran :")
print(f"Le score de santÃ© mentale prÃ©dit est de : {prediction[0]:.2f} / 100")

# 5. ðŸ“ˆ LE GRAPHIQUE (Pour ton dossier)
plt.figure(figsize=(10, 6))
sns.regplot(x='Daily_Usage_Time_min', y='Mental_Health_Index', data=df,
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Impact du temps d\'Ã©cran sur la santÃ© mentale')
plt.xlabel('Temps d\'utilisation quotidien (minutes)')
plt.ylabel('Indice de SantÃ© Mentale')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. On charge ton fichier (attention au nom avec l'espace !)
df = pd.read_csv('social_media_addiction 2.csv')

# 2. On dÃ©finit les variables (X = causes, y = consÃ©quence)
X = df[['Age', 'Daily_Usage_Time_min']]
y = df['Mental_Health_Index']

# 3. On entraÃ®ne l'IA (RÃ©gression LinÃ©aire)
model = LinearRegression()
model.fit(X, y)

# 4. ðŸ”® LA PRÃ‰DICTION
# Test : Un Ã©tudiant de 20 ans qui passe 3h (180 min) sur les rÃ©seaux
test_etudiant = [[20, 180]]
prediction = model.predict(test_etudiant)

print(f"--- RÃ‰SULTAT ---")
print(f"Pour un profil de 20 ans avec 180 min d'Ã©cran :")
print(f"Le score de santÃ© mentale prÃ©dit est de : {prediction[0]:.2f} / 100")

# 5. ðŸ“ˆ LE GRAPHIQUE (Le visuel pour ton jury)
plt.figure(figsize=(10, 6))
sns.regplot(x='Daily_Usage_Time_min', y='Mental_Health_Index', data=df,
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Comment le numÃ©rique "nie" notre santÃ©')
plt.xlabel('Temps d\'utilisation quotidien (minutes)')
plt.ylabel('Indice de SantÃ© Mentale (0 Ã  100)')
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Charger les donnÃ©es
df = pd.read_csv('social_media_addiction 2.csv')

# 2. Variables du modÃ¨le (On prend tout le cÃ´tÃ© toxique)
features = ['Age', 'Daily_Usage_Time_min', 'FOMO_Score', 'Productivity_Loss_Score', 'Scroll_Rate_ppm']
X = df[features]
y = df['Mental_Health_Index']

# 3. L'IA "Random Forest" (Le niveau au-dessus)
model_ml = RandomForestRegressor(n_estimators=100, random_state=42)
model_ml.fit(X, y)

# 4. LE GRAPHIQUE "WOW" : Feature Importance
importances = pd.Series(model_ml.feature_importances_, index=features).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
importances.plot(kind='barh', color='teal')
plt.title('Quels facteurs dÃ©truisent rÃ©ellement la santÃ© mentale ?', fontsize=14)
plt.xlabel('Poids de la variable dans l\'IA')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# 1. PrÃ©paration des donnÃ©es
df = pd.read_csv('social_media_addiction 2.csv')
features = ['Daily_Usage_Time_min', 'Mental_Health_Index', 'Productivity_Loss_Score']
X = df[features]

# On normalise (important pour le clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. L'IA de Clustering (K-Means)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Groupe'] = kmeans.fit_predict(X_scaled)

# 3. Le Graphique "Waw" : Visualisation des profils
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='Daily_Usage_Time_min', y='Mental_Health_Index',
                hue='Groupe', palette='viridis', s=100, alpha=0.7)
plt.title('Segmentation des Comportements NumÃ©riques (K-Means)', fontsize=15)
plt.xlabel('Temps d\'utilisation (min)')
plt.ylabel('SantÃ© Mentale')
plt.legend(title='Profils identifiÃ©s')
plt.show()

# 4. Analyse des profils
print("\n--- ANALYSE DES PROFILS (MOYENNES) ---")
print(df.groupby('Groupe')[features].mean())
```
